```python
np.random.seed(0)


d = 100
n = 100

def func(H, w_star, f_star, w):
    return 0.5 * (w - w_star).T @ H @ (w - w_star) + f_star

def compute_loss(H_list, w_star_list, f_star, w):
    n = len(H_list)
    l = 0.0
    for i in range(n):
        l += func(H_list[i], w_star_list[i], f_star, w)
    return l / n

def compute_grad(H_list, w_star_list, w):
    n = len(H_list)
    g = np.zeros_like(w)
    for i in range(n):
        g += H_list[i] @ (w - w_star_list[i])
    return g / n

def generate_spd_matrix(d, factor=3, seed=0):
    np.random.seed(seed)
    A = np.random.randn(d, factor * d)
    H = A @ A.T 
    return H / (factor * d)

def generate_problem(n, d, f_star=1.0, interpolated=True, seed=0):
    np.random.seed(seed)
    H_list = [generate_spd_matrix(d) for _ in range(n)] 
    if interpolated:
        w_star_shared = np.random.randn(d)
        w_star_list = [w_star_shared for _ in range(n)]
    else:
        w_star_list = [np.random.randn(d) for _ in range(n)]

    return H_list, w_star_list, f_star

no_interpolation_non_zero_f = generate_problem(n, d, f_star=1.0, interpolated=False)
interpolation_non_zero_f = generate_problem(n, d, f_star=1.0, interpolated=True)
interpolation_zero_f = generate_problem(n, d, f_star=0.0, interpolated=True)
```

```python
class SPS(BaseOptimizer):
    
    def __init__(self, params: np.ndarray, 
                 c: float = 0.5, 
                 eps: float = 0.0,
                 f_star: float = 0.0,
                 eta_max: float = np.inf):
        
        self.params = params
        self.lr = 1.0
        self.eps = eps
        self.f_star = f_star
        self.c = c
        self.eta_max = eta_max
        
        self.defaults = dict(
            lr=1.0,
            c=c,
            eps=eps,
            )
        
    def step(self, loss, grad):
        
        self.lr = (loss - self.f_star) / ( self.c * np.linalg.norm(grad)**2 + self.eps )
        self.lr = np.minimum(self.eta_max, self.lr)
        self.params -= self.lr * grad
            
        return loss, grad
```

```python
def train_loop(problem, f_star, iters, eta_max=0.5, eps=1e-10, c=0.5, seed=0):
        
    np.random.seed(seed)
    
    H_list, w_star_list, true_f_star = problem

    w = np.random.randn(d)

    hist = defaultdict(list)

    for i in range(iters):
                
        loss = compute_loss(H_list, w_star_list, true_f_star, w)
        grad = compute_grad(H_list, w_star_list, w)

        lr = (loss - f_star) / ( c * np.linalg.norm(grad)**2 + eps )
        lr = np.minimum(eta_max, lr)
        w = w - lr * grad
        
        
        hist["loss"].append(loss)
        hist["grad_norm_sq"].append(np.linalg.norm(grad)**2)
        hist["lr"].append(lr)
        
    return hist
```

```python
def twin_polyak(problem, iters, eta_max=np.inf, eps=1e-10, seed=0):
        
    np.random.seed(seed)
    
    H_list, w_star_list, true_f_star = problem

    # parameters
    params_x = np.random.randn(d)
    params_y = np.random.randn(d)


    hist = defaultdict(list)

    for i in range(iters):
                
        loss_x, grad_x = compute_loss(H_list, w_star_list, true_f_star, params_x), compute_grad(H_list, w_star_list, params_x)
        loss_y, grad_y  = compute_loss(H_list, w_star_list, true_f_star, params_y), compute_grad(H_list, w_star_list, params_y)
        
        lr_x = np.minimum(( (loss_x - loss_y) / (0.5 * np.linalg.norm(grad_x)**2 + eps) ), eta_max) 
        lr_y = np.minimum(( (loss_y - loss_x) / (0.5 * np.linalg.norm(grad_y)**2 + eps) ), eta_max) 

        # Optimization step
        if loss_x > loss_y:
            params_x = params_x - lr_x * grad_x
            lr = lr_x
        else:
            params_y = params_y - lr_y * grad_y
            lr = lr_y 
        
        if loss_x < loss_y:
            loss, grad, g_norm_sq = loss_x, grad_x, np.linalg.norm(grad_x)**2
        else:
            loss, grad, g_norm_sq = loss_y, grad_y, np.linalg.norm(grad_y)**2
            
        hist["loss"].append(loss)
        hist["grad_norm_sq"].append(g_norm_sq)
        hist["lr"].append(lr)
        hist["lr_x"].append(lr_x)
        hist["lr_y"].append(lr_y)
        
    return hist
```

```python
def dec_sps(problem, f_star, iters, eta_max=1.0, c_0=1.0, eps=1e-10, seed=0):
        
    np.random.seed(seed)
    
    H_list, w_star_list, true_f_star = problem

    w = np.random.randn(d)
    
    hist = defaultdict(list)
    
    c = c_0
    lr = eta_max
    
    for i in range(iters):
                
        loss = compute_loss(H_list, w_star_list, true_f_star, w)
        grad = compute_grad(H_list, w_star_list, w)
        
        c_prev = c
        c = c_0 * np.sqrt(i + 1)
        polyak_lr = (loss - f_star) / ( np.linalg.norm(grad)**2 + eps )
        lr = np.minimum(polyak_lr, c_prev * lr) / c
        w = w - lr * grad
        
        
        hist["loss"].append(loss)
        hist["grad_norm_sq"].append(np.linalg.norm(grad)**2)
        hist["lr"].append(lr)
        
    return hist
```

```python
def moving_average(data, window_width):
    cumsum_vec = np.cumsum(np.insert(data, 0, 0)) 
    ma_vec = (cumsum_vec[window_width:] - cumsum_vec[:-window_width]) / window_width
    return ma_vec

def plotter(histories, labels, colors=None, linestyles=None, 
            linewidths=None, markers=None, markeredgecolor=None, markersize=8,
            f_star=None, suptitle=None, prefix="train",
            x_labels=None, y_labels=None, alphas=None,
            threshold=1e-10, xlims=None, tight_layout=True, filename=None):
    
    fig, ax = plt.subplots(1, 3, figsize=(14, 4))
    if suptitle is not None:
        fig.suptitle(suptitle, fontsize=16)
        
    if colors is None:
        colors = list(plt.cm.colors.XKCD_COLORS.keys())
        np.random.shuffle(colors)
        colors = list(plt.cm.colors.BASE_COLORS) + colors
        
    if linestyles is None:
        linestyles = ["-" for x in histories]
    
    if linewidths is None:
        linewidths = [1.5 for x in histories]
        
    if markers is None:
        markers = [" " for x in histories]
        
    if alphas is None:
        alphas = [1.0 for x in histories]
        
    # if f_star is None:
    #     f_star = np.min([x["loss"] for x in histories])
    # else:
    #     f_star = np.min([f_star for x in histories])
    
    for history, label, c, ls, lw, m, alpha in zip(histories, labels, colors, linestyles, linewidths, markers, alphas):
        f_suboptim = history[f"{prefix}/loss"] - f_star
        
        try:
            markevery = [x + np.random.randint(0, 5) for x in range(0, len(history[f"{prefix}/loss"]) - 5, len(history["test/loss"]) // 10)]
        except:
            markevery = []
        
        ax[0].semilogy(f_suboptim, linestyle=ls, linewidth=lw, color=c, markevery=markevery, marker=m, markeredgecolor=markeredgecolor, markersize=markersize)
        ax[1].semilogy(history[f"{prefix}/grad_norm_sq"], linestyle=ls, linewidth=lw, color=c, label=label, markevery=markevery, marker=m, markeredgecolor=markeredgecolor, markersize=markersize)
        
        try:
            markevery = [0, ] + [x + np.random.randint(0, 5) for x in range(0, len(history["lr"]) - 5, len(history["lr"]) // 10)]
        except:
            markevery = []
        
        if alpha != 1.0:
            ax[2].semilogy(history["lr"], linestyle=ls, linewidth=lw, color=c, alpha=alpha)
            ax[2].semilogy(moving_average(history["lr"], 100), linestyle=ls, linewidth=lw, color=c, alpha=1.0, markevery=markevery, marker=m, markeredgecolor=markeredgecolor, markersize=markersize)
        else:
            ax[2].semilogy(history["lr"], linestyle=ls, linewidth=lw, color=c, markevery=markevery, marker=m, markeredgecolor=markeredgecolor, markersize=markersize)



    if f_star == 0.0:
        ax[0].set_ylabel(r"$f(x_k)$")
        ax[1].set_ylabel(r"$\| f(x_k) \|^2$")
    else:
        ax[0].set_ylabel(r"$(f(x_k) - f^*)$")
        ax[1].set_ylabel(r"$\| f(x_k) \|^2$")
        
    ax[2].set_ylabel("step-size")
    
    ax[0].set_xlabel("Epochs")
    ax[1].set_xlabel("Epochs")
    ax[2].set_xlabel("Steps")


    if xlims is not None:
        ax[1].set_xlim(right=xlims[1])

    fig.legend()
    ax[0].grid()
    ax[1].grid()
    ax[2].grid()

    if tight_layout:
        fig.tight_layout()
        
    if filename is not None:
        directory = "experiments/plots/" + datetime.datetime.today().strftime("%d-%m-%Y")
        if not os.path.exists(directory):
            os.makedirs(directory)
        fig.savefig(directory + "/" + filename)
    else:
        fig.show()    
```

```python
def train_loop(dataset: list[np.ndarray], 
               batch_size: int, 
               n_epochs: int,
               optimizer: BaseOptimizer,
               seed: int = 0,
               **optimizer_kwargs) -> dict: 

    np.random.seed(seed)

    train_data, train_target, test_data, test_target = dataset
    
    # parameters
    params = np.random.randn(train_data.shape[1])
    optim = optimizer(params, **optimizer_kwargs)

    # oracle 
    loss_function = LeastSquaresLoss()
    
    # logging 
    history = defaultdict(list)

    indices = np.arange(train_data.shape[0])
    
    # Train Evaluation 
    loss = loss_function.func(params, train_data, train_target)
    grad = loss_function.grad(params, train_data, train_target)
    
    g_norm_sq = np.linalg.norm(grad)**2
    history["train/loss"].append(loss)
    history["train/grad_norm_sq"].append(g_norm_sq)
    
    # Test Evaluation 
    loss = loss_function.func(params, test_data, test_target)
    grad = loss_function.grad(params, test_data, test_target)
    
    g_norm_sq = np.linalg.norm(grad)**2
    history["test/loss"].append(loss)
    history["test/grad_norm_sq"].append(g_norm_sq)
    
    for epoch in range(n_epochs):
                
        # Training 
        if batch_size != train_data.shape[0]: # If full batch then don't shuffle indices
            np.random.shuffle(indices)

        for idx in range(train_data.shape[0]//batch_size):
            batch_indices = indices[idx*batch_size:(idx+1)*batch_size]
            batch_data = train_data[batch_indices]
            batch_target = train_target[batch_indices] 
        
            train_loss = loss_function.func(params, batch_data, batch_target)
            train_grad = loss_function.grad(params, batch_data, batch_target)

            optim.step(loss=train_loss, grad=train_grad)
            history["lr"].append(optim.lr)

        # Train Evaluation 
        loss = loss_function.func(params, train_data, train_target)
        grad = loss_function.grad(params, train_data, train_target)
        
        g_norm_sq = np.linalg.norm(grad)**2
        history["train/loss"].append(loss)
        history["train/grad_norm_sq"].append(g_norm_sq)
        
        # Test Evaluation 
        loss = loss_function.func(params, test_data, test_target)
        grad = loss_function.grad(params, test_data, test_target)
        
        g_norm_sq = np.linalg.norm(grad)**2
        history["test/loss"].append(loss)
        history["test/grad_norm_sq"].append(g_norm_sq)
    
    return history
```

```python
class SPS(BaseOptimizer):
    
    def __init__(self, params: np.ndarray, 
                 c: float = 0.5, 
                 eps: float = 0.0,
                 f_star: float = 0.0,
                 eta_max: float = np.inf):
        
        self.params = params
        self.lr = 1.0
        self.eps = eps
        self.f_star = f_star
        self.c = c
        self.eta_max = eta_max
        
        self.defaults = dict(
            lr=1.0,
            c=c,
            eps=eps,
            )
        
    def step(self, loss, grad):
        
        self.lr = (loss - self.f_star) / ( self.c * np.linalg.norm(grad)**2 + self.eps )
        self.lr = np.minimum(self.eta_max, self.lr)
        self.params -= self.lr * grad
            
        return loss, grad


class SPS_MA(BaseOptimizer):
    
    def __init__(self, params: np.ndarray, 
                 c: float = 0.5, 
                 eps: float = 1e-8,
                 eta_max: float = 0.0,
                 betas: tuple[float, float] = (0.9, 0.9)):
        
        self.params = params
        self.lr = 1.0
        self.eps = eps
        self.c = c
        self.betas = betas
        self.eta_max = eta_max
        
        self.defaults = dict(
            lr=1.0,
            c=c,
            eps=eps,
            )
        
        self._step_t = 0
        
        self.sum_f = 0.0
        self.sum_g = np.zeros_like(self.params)
        
    def step(self, loss, grad):
        
        self._step_t += 1
        
        self.sum_f = self.betas[0] * self.sum_f + (1 - self.betas[0]) * loss
        self.sum_g = self.betas[1] * self.sum_g + (1 - self.betas[1]) * grad
        # sum_hat = self.sum_g / (1 - self.beta**self._step_t)

        self.lr = self.sum_f / ( self.c * np.linalg.norm(self.sum_g)**2 + self.eps )        
        self.params -= self.lr * grad
            
        return loss, grad
```

```python
def twin_polyak(dataset: list[np.ndarray], 
               batch_size: int, 
               n_epochs: int,
               seed: int = 0,
               ) -> dict: 
    
    np.random.seed(seed)

    train_data, train_target, test_data, test_target = dataset

    # parameters
    params_x = np.random.randn(train_data.shape[1])
    params_y = np.random.randn(train_data.shape[1])

    # oracle 
    loss_function = LeastSquaresLoss()
    
    # logging 
    history = defaultdict(list)

    indices = np.arange(train_data.shape[0])
    
    # Train Evaluation 
    loss, grad = loss_function.func(params_x, train_data, train_target), loss_function.grad(params_x, train_data, train_target)
    g_norm_sq = np.linalg.norm(grad)**2
    history["train/loss"].append(loss)
    history["train/grad_norm_sq"].append(g_norm_sq)
    
    # Test Evaluation 
    loss, grad = loss_function.func(params_x, test_data, test_target), loss_function.grad(params_x, test_data, test_target)
    g_norm_sq = np.linalg.norm(grad)**2
    history["test/loss"].append(loss)
    history["test/grad_norm_sq"].append(g_norm_sq)
    
    eps = 0.0
    
    for epoch in range(n_epochs):
    
        # Training 
        if batch_size != train_data.shape[0]:
            np.random.shuffle(indices)

        for idx in range(train_data.shape[0]//batch_size):
            batch_indices = indices[idx*batch_size:(idx+1)*batch_size]
            
            batch_data = train_data[batch_indices]
            batch_target = train_target[batch_indices] 
            
            loss_x, grad_x = loss_function.func(params_x, batch_data, batch_target), loss_function.grad(params_x, batch_data, batch_target)
            loss_y, grad_y  = loss_function.func(params_y, batch_data, batch_target), loss_function.grad(params_y, batch_data, batch_target)
    
            lr_x = np.minimum(( (loss_x - loss_y) / (0.5 * np.linalg.norm(grad_x)**2 + eps) ), np.inf) 
            lr_y = np.minimum(( (loss_y - loss_x) / (0.5 * np.linalg.norm(grad_y)**2 + eps) ), np.inf) 

            # Optimization step
            if loss_x > loss_y:
                params_x -= lr_x * grad_x
                lr = lr_x
            else:
                params_y -= lr_y * grad_y
                lr = lr_y         
            
            history["lr_x"].append(np.abs(lr_x))
            history["lr_y"].append(np.abs(lr_y))
            history["lr"].append(lr)
            
            
        # Train Evaluation 
        loss_x, grad_x = loss_function.func(params_x, train_data, train_target), loss_function.grad(params_x, train_data, train_target)
        loss_y, grad_y = loss_function.func(params_y, train_data, train_target), loss_function.grad(params_y, train_data, train_target)
        
        if loss_x < loss_y:
            loss, grad, g_norm_sq = loss_x, grad_x, np.linalg.norm(grad_x)**2
        else:
            loss, grad, g_norm_sq = loss_y, grad_y, np.linalg.norm(grad_y)**2
            
        history["train/loss"].append(loss)
        history["train/grad_norm_sq"].append(g_norm_sq)
            
        # Test Evaluation 
        loss_x, grad_x = loss_function.func(params_x, test_data, test_target), loss_function.grad(params_x, test_data, test_target)
        loss_y, grad_y = loss_function.func(params_y, test_data, test_target), loss_function.grad(params_y, test_data, test_target)
        
        if loss_x < loss_y:
            loss, grad, g_norm_sq = loss_x, grad_x, np.linalg.norm(grad_x)**2
        else:
            loss, grad, g_norm_sq = loss_y, grad_y, np.linalg.norm(grad_y)**2
            
        history["test/loss"].append(loss)
        history["test/grad_norm_sq"].append(g_norm_sq)

    return history
```

```python
def twin_model_explicit(dataset: list[np.ndarray], 
               batch_size: int, 
               n_epochs: int,
               seed: int = 0,
               ) -> dict: 
    
    np.random.seed(seed)

    train_data, train_target, test_data, test_target = dataset

    # parameters
    params_x = np.random.randn(train_data.shape[1])
    params_y = np.random.randn(train_data.shape[1])

    # oracle 
    loss_function = LeastSquaresLoss()
    
    # logging 
    history = defaultdict(list)

    indices = np.arange(train_data.shape[0])
    
    # Evaluation 
    loss = loss_function.func(params_x, test_data, test_target)
    grad = loss_function.grad(params_x, test_data, test_target)
    g_norm_sq = np.linalg.norm(grad)**2
    history["test/loss"].append(loss)
    history["test/grad_norm_sq"].append(g_norm_sq)
    
    eps = 0.0
    
    for epoch in range(n_epochs):
    
        # Training 
        if batch_size != train_data.shape[0]:
            np.random.shuffle(indices)

        for idx in range(train_data.shape[0]//batch_size):
            batch_indices = indices[idx*batch_size:(idx+1)*batch_size]
            
            batch_data = train_data[batch_indices]
            batch_target = train_target[batch_indices] 
            
            train_loss_x = loss_function.func(params_x, batch_data, batch_target)
            train_grad_x = loss_function.grad(params_x, batch_data, batch_target) 
            
            train_loss_y = loss_function.func(params_y, batch_data, batch_target)
            train_grad_y = loss_function.grad(params_y, batch_data, batch_target) 
            
            # Optimization step
            if train_loss_x > train_loss_y:
                lr = np.minimum(( (train_loss_x - train_loss_y) / (0.5 * np.linalg.norm(train_grad_x)**2 + eps) ), np.inf) 
                params_x = params_x - lr * train_grad_x
            else:
                lr = np.minimum(( (train_loss_y - train_loss_x) / (0.5 * np.linalg.norm(train_grad_y)**2 + eps) ), np.inf) 
                params_y = params_y - lr * train_grad_y

        history["lr"].append(lr)

        # Evaluation 
        loss_x, grad_x  = loss_function.func(params_x, test_data, test_target), loss_function.grad(params_x, test_data, test_target)
        loss_y, grad_y = loss_function.func(params_y, test_data, test_target), loss_function.grad(params_y, test_data, test_target)
        
        g_norm_sq_x = np.linalg.norm(grad_x)**2
        g_norm_sq_y = np.linalg.norm(grad_y)**2

        
        history["test/loss_x"].append(loss_x)
        history["test/loss_y"].append(loss_y)
        history["test/g_norm_sq_x"].append(g_norm_sq_x)
        history["test/g_norm_sq_y"].append(g_norm_sq_y)
        history["norm_x_y"].append(np.linalg.norm(params_x - params_y)**2)
        
    return history
```
